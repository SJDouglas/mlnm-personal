```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0,1,num=101)
y = np.random.rand(np.size(x)) -.5 + x
```

```python
np.random.rand(3,6) - .5
```

```python
x
```

```python
z = np.polyfit(x, y, 1)
z
m,b = z
m,b
```

```python
plt.scatter(x, y, label='Original data')
#plt.plot(x,z[0]*x + z[1], 'r', label='Fitted line')
plt.plot(x,m*x + b, 'r', label='Fitted line')
plt.legend()
plt.show()
```
